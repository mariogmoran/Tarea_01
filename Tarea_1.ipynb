Ejercicio_9 FACTORIAL DE UN NMERO

P9.jpg

```
print("Este programa te devuelve el nmero factorial de un nmero entero")
num=int(input("Ingrese un nmero entero positivo: "))
factorial=1
for i in range(num):
  factorial=factorial*num
  num-=1
print("El nmero factorial es:",factorial)
print("Fin del programa")
```

```
print("Este programa calcula el rea de un crculo, y con el resultado puede calcular el volmen de un cilindro")
print("rea de un crculo")
Pi=3.14159265
print("Ingrese el radio de la circunferencia") 
r=int(input()) 
A=Pi*(r**2)
print("El rea del crculo es:",A)

print("Volmen de un cilindro")
print("Ingrese la altura del cilindro")
h=int(input()) 
print("El volumen del cilindro es:",A*h)
```
